# Visulization

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='State_name', y='Population', data=df.sort_values(by='Population', ascending=False).head(10))
plt.xticks(rotation=45)
plt.title('Top 10 States by Population')
plt.xlabel('State Name')
plt.ylabel('Population')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.pie(df['State_name'].value_counts().head(10), labels=df['State_name'].value_counts().head(10).index, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 States by Population Distribution')
plt.axis('equal')
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.countplot(x='State_name', data=df, order=df['State_name'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Count of Records by State')
plt.xlabel('State Name')
plt.ylabel('Count')
plt.show()
```

```python
religion_cols = ['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains']
religion_totals = df[religion_cols].sum()

plt.figure(figsize=(10, 6))
sns.barplot(x=religion_totals.index, y=religion_totals.values, palette='Set2')
plt.xlabel('Religion')
plt.ylabel('Total Population')
plt.title('Total Population by Religion in India Census 2011')
plt.show()
```
